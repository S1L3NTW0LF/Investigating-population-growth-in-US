### Data Cleaning
 

I will now transpose the table so that the states are listed as rows and the columns are the state characteristics.


```python
cdt = cd.transpose()
cdt.head()
```

```python
cdt.shape
```

There are 51 rows which are the states and there are 85 columns which are the state charactersitics. There is one additional 
row because there is the Fact Note row. Since I only need the states and a few of the state characteristics I will remove Fact Note and all state characteristics not related to the questions I am trying to answer. I will also rename the columns so that they are not so long.

```python
ccd = cdt.iloc[1:51, [0, 2, 21, 47]]
ccd.rename(columns={'Population estimates, July 1, 2016,  (V2016)': 'Population', 'Population, percent change - April 1, 2010 (estimates base) to July 1, 2016,  (V2016)': 'Population % change', 'Foreign born persons, percent, 2011-2015': 'Foreign born %', 'Median household income (in 2015 dollars), 2011-2015': 'Household income'}, inplace=True)
ccd.head()
```

I will now change the datatypes to integers or floats so that I can perform calculations on them.

```python
def transform(col):
    col = col.str.replace(',', '').str.replace('$', '')
    col = pd.to_numeric(col)
    return col
```

```python
def to_percent(col):
    col = col.str.replace('%', '')
    col = pd.to_numeric(col)
    col = col / 100
    return col
```

```python
ccd['Population'] = transform(ccd["Population"])
ccd['Household income'] = transform(ccd["Household income"])
ccd['Population % change'] = to_percent(ccd["Population % change"])
ccd['Foreign born %'] = to_percent(ccd["Foreign born %"])
ccd.info()
```

The data has now been cleaned. The rows now contain only states. The columns contain state characteristics and the datatypes are integers and floats.

## Exploratory Data Analysis


### Research Question 1 
### (Are states with smaller populations growing faster than those with larger populations by percent?)

First to get a sense of the data I will look over basic statistics for each column.

```python
ccd.describe()
```

>For population it looks like the mean is around 6.5 million and it ranges from around 600,000 to 40,000,000. 

>For Population % change the mean is 3.12% and the range is -1.2% to 10.8%.

>For Foreign born % the mean is 7.1% and the range is 0.03% to 27.0%.

>For Household income the mean is around 54,000 dollars and the range is about 40,000 to 75,000 dollars.

```python
ccd.hist(figsize=(15,10))
plt.show()
```

> Population is skewed right meanining that there are more states with smaller populations than larger populations. 

> Population % change does not show an obvious distribution. 

> This shows that the number of states with a foreign born % population is skewed right meaning that there are not that many states with a high population of foreigners. 

> There does not seem to be an obvious distribution for Household income. The median does look to be around 50,000 dollars.

```python
x = ccd['Population']
y = ccd['Population % change']
plt.scatter(x, y)
plt.xlabel('Population')
plt.ylabel('Population % change')
plt.title('Population vs. Population % change')
plt.show()
```

There is a weak correlation between Population of a state and the Population % change. Otherwise the correlation looks to be positive. From this graph it is also obvious how many states have smaller populations.

### Research Question 2 
#### (How does the % of foreign born population and median household income correlate to population growth?)

```python
x = ccd['Foreign born %']
y = ccd['Population % change']
plt.scatter(x, y)
plt.xlabel('Foreign born %')
plt.ylabel('Population % change')
plt.title('Foreign born % vs. Population % change')
plt.show()
```

There is a positive correlation between Population % change and Foreign born %. There is a cluster in the lower left of the scatter plot and it slowly expands outwards in a cone like shape avoiding the edges of the scatterplot.

```python
x = ccd['Household income']
y = ccd['Population % change']
plt.scatter(x, y)
plt.xlabel('Household income')
plt.ylabel('Population % change')
plt.title('Household income vs. Population % change')
plt.show()
```

There is a weak correlation or no correlation. It could be said states with a median household incomes that are more average have higher population growth. This is based off of the fact that there are no states at the top left or top right corner of the scatter plot.
